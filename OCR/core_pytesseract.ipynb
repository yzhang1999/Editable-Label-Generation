```python
import os
from PIL import Image
import pytesseract

# Input folder containing the images
image_folder = '/Users/darwinye/myfile/NorthwesternU/499 Capstone/Data_subset_Final/test/Images'
# Output folder to save the text files
output_folder = '/Users/darwinye/myfile/NorthwesternU/499 Capstone/Data_subset_Final/test/OCR_text'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Loop through all files in the image folder
for file_name in os.listdir(image_folder):
    # Get the full path of the image file
    file_path = os.path.join(image_folder, file_name)
    
    # Check if the file is an image
    if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        try:
            # Open the image
            image = Image.open(file_path)
            
            # Perform OCR to detect text in the image
            detected_text = pytesseract.image_to_string(image)
            
            # Define the output text file name (same as the image name with .txt extension)
            text_file_name = os.path.splitext(file_name)[0] + ".txt"
            text_file_path = os.path.join(output_folder, text_file_name)
            
            # Save the detected text to the text file
            with open(text_file_path, "w") as text_file:
                text_file.write(detected_text)
            
            print(f"Processed and saved: {text_file_name}")
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
```
